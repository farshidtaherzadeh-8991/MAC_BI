```python
csv_files = glob.glob("Data_projet_csv/*.csv")

for file in csv_files:
    try:
        # tashkhis delimiter
        with open(file, "r", encoding="utf-8") as f:
            sample = f.read(2048)
            dialect = csv.Sniffer().sniff(sample)
            delimiter = dialect.delimiter

        # esm table ro az esm file migirim
        table_name = file.split("/")[-1].replace(".csv", "").lower()

        # khondan csv ba delimiter dorost
        df = pd.read_csv(file, sep=delimiter, on_bad_lines="skip")

        # ersal be postgres
        df.to_sql(table_name, engine, schema="raw_mac_f", if_exists="replace", index=False)

        print(f"{file} → imported to table raw_mac_f.{table_name}")

    except Exception as e:
        print(f"Error importing {file}: {e}")
```
